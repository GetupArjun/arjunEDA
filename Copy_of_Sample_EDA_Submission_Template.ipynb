## 3. ***Data Wrangling***

### Data Wrangling Code

## Columns like Agent and Company have maximum number of null values as compared to other columns and we will replace all null values with 0, because these are not missing values instead they will be considered as "Not Applicable".

````python
# Write your code to make your dataset analysis ready.

# Let's handle null values.
# Replacing null values of column "Agent" and "Company" with 0

Hotel_Bookings[['agent', 'company']] = Hotel_Bookings[['agent', 'company']].fillna(0.0)
````

````python
Hotel_Bookings.isna().sum().sort_values(ascending=False)
````

## Replace missing values by mode and mean

````python
for i in Hotel_Bookings.columns:
    if(Hotel_Bookings[i].dtype=="object"):
        replace=Hotel_Bookings[i].mode()[0]
        Hotel_Bookings[i]=Hotel_Bookings[i].fillna(replace)
    else:
        replace=Hotel_Bookings[i].mean()
        Hotel_Bookings[i]=Hotel_Bookings[i].fillna(replace)
````

````python
Hotel_Bookings.isna().sum().sort_values()
````

# If the column dtype=objective then missing values replace by mode.

#### Chart - 15 - Pair Plot 

````python
# Pair Plot visualization code
#In Bivariate Analysis let's take our 'Y' axis will be "lead_time  "

plt.figure(figsize=(30,120))
w=1
for i in Hotel_Bookings.columns:
    if(Hotel_Bookings[i].dtype=="object"):
        plt.subplot(11,3,w)
        sb.boxplot(Hotel_Bookings[i],Hotel_Bookings["lead_time"])
        w=w + 1
        
    else:
        plt.subplot(11,3,w)
        sb.scatterplot(Hotel_Bookings[i],Hotel_Bookings["lead_time"])
        w=w + 1
````

##### 1. Why did you pick the specific chart?

## Bivariate Analysis for every two columns (x,y) which is given in our data set.
